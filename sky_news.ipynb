```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin

# URL of the Sky News homepage
base_url = "https://news.sky.com/"

# Send a GET request to the webpage
response = requests.get(base_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    sky_news = {}
    titles_list = []
    content_list = []
    sections = soup.find_all('section', class_="wrap gap-400")

    for section in sections:
        title = section.find('span', class_="ui-section-header-title").get_text(strip=True)
        titles_list.append(title)
        cels = section.find_all('div', class_='grid-cell')
        contents = []
        for cel in cels:
            content = cel.find('div', class_='ui-story-headline').get_text(strip=True)
            link_tag = cel.find('div', class_='ui-story-headline').find('a')
            if link_tag:
                link = urljoin(base_url, link_tag.get('href'))
                
                content_with_link = f"{content} '{link}'"
                contents.append(content_with_link)
        content_list.append(contents)

    # Prepare the data for the DataFrame
    data = {title: contents for title, contents in zip(titles_list, content_list)}

    # Convert to DataFrame
    df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in data.items()]))

    print(df)
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")

df
```

```python
df.to_csv('sky_news2.csv')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin
from sqlalchemy import create_engine

# URL of the Sky News homepage
base_url = "https://news.sky.com/"

# Send a GET request to the webpage
response = requests.get(base_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    sky_news = {}
    titles_list = []
    content_list = []
    sections = soup.find_all('section', class_="wrap gap-400")

    for section in sections:
        title = section.find('span', class_="ui-section-header-title").get_text(strip=True)
        print(title)
        titles_list.append(title)
        cels = section.find_all('div', class_='grid-cell')
        contents = []
        for cel in cels:
            content = cel.find('div', class_='ui-story-headline').get_text(strip=True)
            link_tag = cel.find('div', class_='ui-story-headline').find('a')
            if link_tag:
                link = urljoin(base_url, link_tag.get('href'))
                content_with_link = f"{content} ({link})"
                print(content_with_link)
                contents.append(content_with_link)
        content_list.append(contents)
        sky_news[title] = contents

    # Convert to DataFrame
    df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in sky_news.items()]))
    print(df)

    # SQLAlchemy connection string
    engine = create_engine('mysql+pymysql://root:@localhost/sky_news')

    # Write DataFrame to MySQL
    df.to_sql(con=engine, name='table_name_for_df', if_exists='replace', index=False)

    print("DataFrame written to MySQL database.")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")


```
